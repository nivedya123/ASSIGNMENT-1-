QUESTION : 

Plot using sns.relplot, line plot, box plot on fmri data set



```python
#importing libraries

import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns

print("Successfully installed...")
```

```python
# importing dataset

fmri=sns.load_dataset('fmri')
fmri.head(5)
```

```python
#rel plot
plt.title("RELATIONAL PLOT")
sns.relplot(data=fmri, x="timepoint", y="signal", hue="region")
plt.show()
```

```python
#line plot

plt.title("LINE PLOT")
sns.lineplot(data=fmri,x='subject',y='signal')
plt.show()
```

```python
#box plot
plt.title("BOX PLOT")
sns.boxplot(data=fmri,x='timepoint',y='signal')
plt.show()
```
